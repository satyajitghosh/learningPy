### Sorting Values in a Pandas Data Frame

```python
#Sort Ascending
orders.Sales.sort_values()
```

```python
orders.Sales.sort_values(ascending=False).head()
```

```python
# Sorting columns whose names have spaces.
orders['Order Date'].sort_values().head()
```

```python
#Sorting the data inside the data frame, not just printing the column in the sorted format
orders.sort_values('Sales').head()
```

```python
# Sorting with two columns
orders.sort_values(['Sales','Discount'],ascending=False).head()
```

### Applying Filters to a data frame

```python
# Filtering numeric data 
orders[orders.Sales > 10000].head()
```

```python
# Filtering char data 
orders[orders['Order Priority'] != 'High'].head()
```

```python
#Filetring with multiple criteria
orders[(orders['Order Priority'] != 'High') & (orders['Customer Segment'] == 'Small Business')].head()
```

```python
orders[(orders['Order Priority'] != 'High') | (orders['Customer Segment'] == 'Small Business')].head()
```

```python
orders[(orders['Order Priority'] != 'High') | (orders['Order Priority'] == 'Low')].head()
```

```python
orders[(orders['Order Priority'] != 'High') & (orders['Order Priority'] != 'Low')].head()
```

### Using isin to shorten OR in filter 

```python
orders[orders['Ship Mode'].isin(['Regular Air','Delivery Truck'])].head()
```

### Deleting rows and columns

```python
orders_del = orders.copy()
```

```python
orders_del.head(2)
```

```python
## axis = 1 implies we are dropping a column
## if inplace = False or not specified, the columns will be dropped while printing output, but it will not be dropped from the DF.

orders_del.drop('Row ID',axis = 1,inplace = True)
```

```python
orders_del.head(2)
```

```python
## Dropping more than one columns
orders_del.drop(['Order Date','Order Priority'],axis = 1,inplace=True)
```

```python
orders_del.head()
```

```python
## Dropping using index
orders_del.drop([2,3],axis=0,inplace=True)
```

```python
## Notice - rows with index 2,3 have been removed. 0,1,4,5..
orders_del.head()
```
