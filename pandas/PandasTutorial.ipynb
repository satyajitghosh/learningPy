```python
iris = pd.read_csv('data/iris.csv')
```

```python
iris.head()
```

```python
#iterating over columns
# Key is the column name and the value is the content of the column.
# The output key printed in each iteration is a name and out value printed in each iteration is a data frame(shape - m x 1).
for key,value in iris.head(3).iteritems():
    print(key,value)
```

```python
## It is important to understand how this works.
## In the previous case,the output key printed in each iteration is a name and out value printed in each iteration is a data frame.
## Therfore,when we print value[0], the first row (1x1) of the value data frame(mx1) is printed in each iteration.
for key,value in iris.head(3).iteritems():
    print(key,value[0])
```
